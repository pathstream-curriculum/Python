## Hypothesis Test Activity

**Goal:** In this activity, you will perform a hypothesis test to determine whether there is a significant difference in revenue during promotions weeks when compared to non-promotion weeks. 

Start by importing any Python libraries or specific functions you will need to complete this activity. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.weightstats import ttest_ind
from statsmodels.stats.proportion import proportions_ztest

sns.set()
```

## 1. Import the data

Import the retail data from the file `retail_promotions_seattle_04.csv`. 

```python
retail_promo_df = pd.read_csv('retail_promotions_seattle_04.csv')
```

## 2. Visualize the Data

Create a violinplot or boxplot to compare the distributions of Weekly_Sales during promotion and non-promotion weeks. 

```python
fig = plt.figure()
sns.violinplot(data=retail_promo_df, x="Weekly_Sales", y="Promo_Week", orient='h')
fig.suptitle("Weekly Sales Distribution, Seattle")
```

## 3. Run a t-test of Independence

Split the data into promotional weeks and non-promotional weeks. Then, run a t-test of independence to determine whether there is a significant difference in weekly sales between these two categories. 

```python
# split the data
promo_wk_sales = retail_promo_df.loc[retail_promo_df["Promo_Week"]==True]["Weekly_Sales"]
non_promo_wk_sales = retail_promo_df.loc[retail_promo_df["Promo_Week"]==False]["Weekly_Sales"]
```

```python
# pass the data to the ttest_ind function
ttest_ind(promo_wk_sales, non_promo_wk_sales)
```
