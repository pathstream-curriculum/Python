```python
import matplotlib.pyplot as plt

plt.style.use('default')
```

```python
will_vote = [82, 64]
totals = [100, 100]
labels = ["Has Children", "No Children"]
```

```python
plt.grid(axis='y', zorder=1)
plt.bar(labels, totals, color='grey', zorder=2)
plt.bar(labels, will_vote, color=['tab:blue', 'tab:orange'], zorder=2)
plt.ylabel('Count')
plt.title("Do you intend to vote in the upcoming election?")
plt.text(-.22, 71, "Yes: 82", color="#cccccc", fontdict={"weight":"bold", "size":15 }, zorder=3)
plt.text(-.2, 86, "No: 18", color="#cccccc", fontdict={"weight":"bold", "size":15 }, zorder=3)
plt.text(0.8, 54, "Yes: 64", color="#444444", fontdict={"weight":"bold", "size":15 }, zorder=3)
plt.text(0.82, 69, "No: 36", color="#444444", fontdict={"weight":"bold", "size":15 }, zorder=3)
```
