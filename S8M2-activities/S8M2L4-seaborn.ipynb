## Import pandas, pyplot, and seaborn
set the seaborn style

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

## Import the data

In this activity you will work with data from all four sites for the year 2013

filename: 'pollution_ca_2013.csv'


```python
pollution_df = pd.read_csv('pollution_ca_2013.csv')
```

```python
pollution_df.head()
```

## Single violinplot

Create a violinplot for the distribution of "CO Mean" acroess the entire database

Add a title!

```python
fig = plt.figure()
ax = sns.violinplot(y="CO Mean", data=pollution_df)
ax.set_title("Distribution of CO2 Measurements")
```

## Group of violinplots
 Create a group of violinplots showing the distribution of "SO2 Mean", separated by site number

```python
fig = plt.figure()
ax = sns.violinplot(x="Site Num", y="CO Mean", data=pollution_df)
ax.set_title("Distribution of CO2 Measurements Across Sites")
```

## Swarmplots
Create a swarmplot for the "SO3 Mean" measure, split by site number.
set the figsize to (8, 10)

```python
fig = plt.figure(figsize=(8,10))
ax = sns.swarmplot(x="Site Num", y="SO2 Mean", data=pollution_df)
```

## Pair Plots

filter pollution_df down to the following columns (create a new df): 
* City
* NO2 Mean
* O3 Mean
* SO2 Mean
* CO Mean

Then, create a pairplot from this dataframe

```python
columns_to_keep = ["City", "NO2 Mean", "O3 Mean", "SO2 Mean", "CO Mean"]
pollution_df_filt = pollution_df[columns_to_keep]
```

```python
fig = plt.figure()
ax = sns.pairplot(pollution_df_filt)
```
